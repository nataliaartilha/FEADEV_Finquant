```python
#importando bibliotecas
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

1.Ações Credit Suisse

```python
credit_suisse = yf.download('CSGN.SW', start = '2023-01-01', end = '2023-03-22', progress = False)
credit_suisse.head()
```

```python
credit_suisse["Adj Close"].plot(figsize=(15, 10))
plt.show()
```

2.Ações Credit Suisse e Bancos brasileiros

```python
df_BR= pd.DataFrame()
bank_br=['CSGN.SW','BBAS3.SA','BBDC4.SA','ITUB4.SA','SANB11.SA','BPAC11.SA','CTGP34.SA','BOAC34.SA']
```

```python

for i in bank_br:
    df_BR[i] = yf.download(i, start = '2023-03-10', end = '2023-03-23', progress = False)['Adj Close']
df_BR.head()
```

```python
#normalizando os preços pro dia primeiro representar 100%
(df_BR/df_BR.iloc[0]*100).plot(figsize=(16,8))
```

```python
#annot pôr numero nos quadradinhos
sns.heatmap(df_BR.corr(),vmin=-1,vmax=1, annot=True,cmap='BrBG')
```
